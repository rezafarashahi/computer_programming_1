In this code example we want to plot y=sin(x) for x between 0 and pi

### **Step 1: Import Required Libraries**
First, we need to import the required libraries:
- `numpy` for numerical operations.
- `matplotlib.pyplot` for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,1,2], [4, 5, 6])
```

### **Step 2: Define the x-values (Input Range)**
Since we want to plot a sine wave from \(0\) to \(pi), we create an array of values between these limits using `numpy.linspace()`.

```python
x = np.linspace(0, 3.14, 100) #this means create an array of numbers for me that starts with 0 ends with 3.14 with 100 points in between.
print(x)
```

 `np.linspace(start, stop, num)` creates `num` evenly spaced values between `start` and `stop`.

---

### **Step 3: Compute the y-values (sin values)**
Now, we compute the sine of each `x` value using `numpy.sin()`.

```python
y = np.sin(x)  # Compute sine for each x-value
```

---

### **Step 4: Plot the Sine Wave**
We use the `plt.plot()` function to plot the sine wave.

```python
plt.plot(x, y)
```
